```python
import random

class WumpusWorld:
    def __init__(self, rows=4, cols=3):
        self.rows = rows
        self.cols = cols
        self.grid = [[' ' for _ in range(cols)] for _ in range(rows)]
        self.wumpus_location = (3, 1)  # Wumpus location (row, column)
        self.gold_location = (2, 2)    # Gold location (row, column)
        self.pit_locations = [self.generate_random_location(exclude=[self.wumpus_location, self.gold_location])]
        self.agent_location = (0, 0)
        self.game_over = False

    def generate_random_location(self, exclude=[]):
        location = (random.randint(0, self.rows-1), random.randint(0, self.cols-1))
        while location in exclude:
            location = (random.randint(0, self.rows-1), random.randint(0, self.cols-1))
        return location

    def print_grid(self):
        for row in self.grid:
            print('|'.join(row))
            print('-' * self.cols * 2 + '-')

    def update_grid(self):
        self.grid = [[' ' for _ in range(self.cols)] for _ in range(self.rows)]
        x, y = self.agent_location
        self.grid[y][x] = 'A'
        if self.agent_location == self.wumpus_location:
            self.game_over = True
            print("You've been eaten by the Wumpus! Game Over.")
        elif self.agent_location == self.gold_location:
            self.game_over = True
            print("Congratulations! You've found the gold! You win!")
        elif self.agent_location in self.pit_locations:
            self.game_over = True
            print("You fell into a pit! Game Over.")
        else:
            if self.wumpus_location[0] == x and abs(self.wumpus_location[1] - y) <= 1:
                print("You smell a Wumpus nearby!")
            if self.wumpus_location[1] == y and abs(self.wumpus_location[0] - x) <= 1:
                print("You smell a Wumpus nearby!")
            for pit_loc in self.pit_locations:
                if pit_loc[0] == x and abs(pit_loc[1] - y) <= 1:
                    print("You feel a draft from a nearby pit!")
                if pit_loc[1] == y and abs(pit_loc[0] - x) <= 1:
                    print("You feel a draft from a nearby pit!")

    def move_agent(self, direction):
        x, y = self.agent_location
        if direction == 'up' and y > 0:
            self.agent_location = (x, y - 1)
        elif direction == 'down' and y < self.rows - 1:
            self.agent_location = (x, y + 1)
        elif direction == 'left' and x > 0:
            self.agent_location = (x - 1, y)
        elif direction == 'right' and x < self.cols - 1:
            self.agent_location = (x + 1, y)
        else:
            print("Invalid move! Try again.")

    def play(self):
        while not self.game_over:
            self.update_grid()
            self.print_grid()
            move = input("Enter your move (up/down/left/right): ").lower()
            self.move_agent(move)

if __name__ == "__main__":
    game = WumpusWorld(rows=4, cols=3)
    game.play()

```
